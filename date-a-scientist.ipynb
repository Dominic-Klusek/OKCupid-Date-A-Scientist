<h1>Data Cleaning</h1>
<p>Removal of nan values, and conversion of any data types as necessary</p>

```python
# get some background info
for column in dating_profiles.columns:
    if('essay' in column):
        continue
    print(dating_profiles[column].value_counts(dropna=False))
    print()
    print()
```

```python
# remove all entries with nan
dating_profiles = dating_profiles.dropna()
dating_profiles = dating_profiles.dropna()

# simplify sign data by removing superfluous data
def sign_return(string):
    # extract sign from data and return
    try:
            if(type(string) == float):
                return 'child of universe'
            for sign in ['leo','libra','cancer','virgo','scorpio','gemini','taurus','aries','pisces','aquarius','sagittarius','capricorn']:
                if(sign in string):
                    return sign
    except:
        return sign
        
dating_profiles['sign'] = dating_profiles['sign'].apply(sign_return)


# simplify education by removing superfluous data
def return_education(string):
    if('dropped out ' in string):
        return 'drop out'
    elif('working on ' in string):
        return 'currently attending'
    else:
        return string.replace('graduated from ', '')
    return None

dating_profiles['education'] = dating_profiles['education'].apply(return_education)


# simplify religion by removing superfluous data
def return_religion(string):
    # extract sign from data and return
    try:
        for rel in ['agnosticism','other','atheism','christianity','catholicism','judaism','buddhism','islam','hinduism']:
            if(rel in string):
                return rel
    except:
        return rel

dating_profiles['religion'] = dating_profiles['religion'].apply(return_religion)


def return_ethnicity(string):
    if (string in ['white','asian','hispanic / latin','black','other']):
        return string
    else:
        return 'other'

dating_profiles['ethnicity'] = dating_profiles['ethnicity'].apply(return_ethnicity)


# simplify diet
def return_diet(string):
    try:
        if (type(string) is 'float'):
            return 'anything'
        for diet in ['anything','vegetarian','vegan','halal','kosher']:
            if(diet in string):
                return diet
    except:
        return string

dating_profiles['diet'] = dating_profiles['diet'].apply(return_diet)
#dating_profiles['diet'] = dating_profiles['diet'].replace({'anything':0,'vegetarian':1,'vegan':2,'halal':3,'kosher':4})
```

```python
# use ordinal encoding to convert string to numeral
for data_label in ['sign', 'education', 'religion', 'ethnicity', 'diet', 'body_type', 'job', 'pets', 'sex', 'smokes', 'status', 'orientation']:
    encoder = OrdinalEncoder()
    encoder.fit(np.array(dating_profiles[data_label].values).reshape(-1,1))
    dating_profiles[data_label] = encoder.transform(np.array(dating_profiles[data_label].values).reshape(-1,1))
```

```python
# get some background info
for column in dating_profiles.columns:
    if('essay' in column):
        continue
    print(dating_profiles[column].value_counts(dropna=False))
    print()
    print()
```
